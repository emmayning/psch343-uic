## 2. Course Overview

Welcome to PSCH 343, Statistical Methods in Behavioral Science! The goal of this class is to equip you with the **essential analytical mindset and tools to interpret and evaluate quantitative research findings**. Building on the knowledge of experimental design (PSCH 242), research process, and research writing, we will explore how statistical methods inform psychological research and data-driven decision-making.

Topics covered include: 1). descriptive statistics (mean, standard deviation); 2). sampling, and 3). inferential statistics (t-tests, ANOVA, correlation). These concepts form the foundation of psychological research methods and data science. **All course material will be motivated by practical applications of these tools in the context of research in both academia and industry**. 

After this class, you should be ready to contribute to a research team from day one. You will be able to summarize data, run basic statistical tests, and present findings clearly. Whether informing research decisions or setting the stage for more advanced analyses, you will have the skills to translate raw data into meaningful insights that drive the next steps in a study.

## 3. Learning Objectives

* **Summarize and visualize** data using descriptive statistics and graphs, and explain observed patterns in clear, non-technical language.
* **Translate** hypotheses into testable statistical analyses.
* **Select and justify** appropriate statistical tests for a given hypothesis and study design.
* **Execute** statistical analyses from start to finish â€” carrying out the chosen test accurately by hand and with open-source software, and articulating the reasoning for each step.
* **Interpret results and communicate findings** effectively for different audiences and decision-making contexts.
